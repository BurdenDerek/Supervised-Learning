```python
import pandas as pd
from path import Path

file_path = Path("Resources/loans_data.csv")
df = pd.read_csv(file_path)
df.head()
```

```python
# binary encoding using pandas single column
laons_binary_encoded = pd.get_dummies(df, columns=['gender'])
laons_binary_encoded.head()
```

```python
# binary encoding using pandas multiple columns
loans_binary_encoded = pd.get_dummies(df, columns=['amount', 'term', 'month', 'age', 'education', 'gender'])
loans_binary_encoded.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df2 = df.copy()
df2["education"] = le.fit_transform(df["education"])
df2.head()
```

```python
df2["gender"] = le.fit_transform(df["gender"])
df2
```

```python
# custom encoding
loan_df = df.copy()
label_encoder = LabelEncoder()
loans_df['month_le'] = label_encoder.fit_transform(loan_df['month'])
loans_df
```

```python
months_num = {
   "January": 1,
   "February": 2,
   "March": 3,
   "April": 4,
   "May": 5,
   "June": 6,
   "July": 7,
   "August": 8,
   "September": 9,
   "October": 10,
   "November": 11,
   "December": 12,
}
```

```python
loans_df["month_num"] = loans_df["month"].apply(lambda x: months_num[x])
loans_df
```
